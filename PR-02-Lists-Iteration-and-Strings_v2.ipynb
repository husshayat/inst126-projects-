````python
def word_redacter(words, top_secret_words):
    """
    Redacting special words from lists 
    
    parameters 
    words: list
    top_secret_words: list that will be redacted
    
    returns
    returns the list of new words displaying the secret words as "REDACTED"
    """
    redacted_count = 0 #counting redacted words
    
    for word in words: #iteration
        if word in top_secret_words: #if the words are in the top_secret_word list
            print("REDACTED") #prints the word "REDACTED"
            redacted_count += 1 #adds 
        else:
            print(word) #if its not a top_secret_word the actual word prints
    print(f"\nRedacted {redacted_count} {'word' if redacted_count== 1 else 'words'}.") #creates a statement of how many words are redacted
    return(words, top_secret_words) #returns words and top_secret_words that are redacted
````

### The problem

Write a program that takes a list of emails, and a list of blocked domains (like umd.edu, NOT the whole email), and prints out only &ldquo;cleaned&rdquo; emails (one e-mail per line) that are not from the blocked domains, as well as the number of times any email from a blocked domain showed up.

### Example input

For example, given these emails and blocked domains:
```python

  "joelchan@umd.edu",
  "firebrand@gmail.com",
  "funlove@yahoo.com",
  "miro@outlook.com",
  "sp@umd.edu",
  "kjfrog@gmail.com"          
]

```

your program should produce the following output:

```

EMAIL: firebrand@gmail.com
EMAIL: sp@umd.edu
EMAIL: kjfrog@gmail.com

Filtered 2 emails from blocked domains.
```

‚¨áÔ∏è‚¨áÔ∏è‚¨áÔ∏è Enter your solution in the cell below ‚¨áÔ∏è‚¨áÔ∏è‚¨áÔ∏è

````python
def email_blocker (emails, blocked_domains):
    """
    blocks emails from printing if domain is blocked 
    
    parameters
    emails: list of emails to be filtered
    blocked_domains: list of domains that are blocked 
    
    return
    returns cleaned emails and number of blocked emails
    """
    clean_emails = [] #empty list to store emails
    num_blocked_emails = 0 #sets the count for blocked emails
    
    for email in emails: #iterating through email list
        domain = email.split('@')[-1] #taking each domain from the email
        if domain not in blocked_domains: 
            clean_emails.append(email) #emails are added to clean list if not blocked
        else:
            num_blocked_emails +=1 #count of blocked emails go up one once detected
    
    for clean_email in clean_emails:
        print(f"EMAIL: {clean_email}") #prints list of clean emails 
        
    num_blocked_emails = len(emails)- len(clean_emails) #calculation for num of blocked emails
    print(f"\nFiltered {num_blocked_emails} emails from blocked domains.") #prints number of blocked emails form blocked domains
    return (clean_emails)
    
````
